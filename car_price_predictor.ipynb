```python
df1 = pd.read_csv('https://raw.githubusercontent.com/the-learning-machine/data/master/tlm_project1/Automobile_data1.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/the-learning-machine/data/master/tlm_project1/Automobile_data2.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/the-learning-machine/data/master/tlm_project1/Automobile_data3.csv')
```

```python
df3
```

```python
print("Data shape of df1 is (X, Y)",df1.shape)
print("Data shape of df2 is (X, Y)",df2.shape)
print("Data shape of df3 is (X, Y)",df3.shape)
```

```python
df = pd.concat([df1, df2],ignore_index=True)
res_df = pd.concat([df, df3],ignore_index =True)
```

```python
res_df.shape
```

```python
res_df
```

```python
def uni(col):
    uniq = []
    for i in range(0,204):
        #print(res_df[col][i])
        if res_df[col][i] not in uniq:
            uniq.append(res_df[col][i])
        else:
            continue
    return uniq
for i in res_df.columns:
    print(i, uni(i))
    print()
```

```python
ncols = []
cols = []
rescols = []
for i in res_df.columns:
    if res_df[i].isnull().sum() != 0:
        ncols.append(i)
for i in res_df.columns:
    cols.append(i)
for i in cols:
    if i not in ncols:
        rescols.append(i)   

nfloat = []
nobj = []
for i in ncols:
    if res_df[i].dtype == 'float64':
        nfloat.append(i)
    else:
        nobj.append(i)
for i in nfloat:
    res_df[i].fillna(0, inplace = True)
for i in nobj:
    res_df[i].fillna('None', inplace = True)
res_df.isnull().sum()

exc = ['?','!']
for i in cols:
    for j in range(0,204):
        if res_df[i][j] in exc:
            res_df[i][j] = 'None'

lis = []
for i in cols:
    for j in range(0,204):
        if res_df[i][j] in exc:
            lis.append(i)
lis
```

```python
for i in range(len(res_df)):
    res_df['price'][i] = res_df['price'][i] / 1000
```

```python
res_df['price'].unique()
```

```python
for i in range(len(res_df)):
    res_df['peak-rpm'][i] = float(res_df['peak-rpm'][i])
```

```python
for i in range(len(res_df)):
    if res_df['stroke'][i] == 'None':
        res_df['stroke'][i] = 0
```

```python
res_df['stroke'] = res_df['stroke'].astype('float64')
res_df.dtypes
```

```python
for i in res_df.columns:
    print(i, ':', res_df[i].unique(), res_df[i].dtype)
```

```python
avg_make = res_df.groupby('make').agg({'price': np.mean, 'engine-size': np.mean, 'bore': np.mean, 'stroke': np.mean, 'peak-rpm': np.mean})
avg_make
```

```python
styles = res_df.groupby(['body-style','make']).agg({'price': np.mean, 'engine-size': np.mean, 'bore' : np.mean, 'stroke': np.mean})
styles
```

```python
total_price = res_df.groupby('make').agg({'price' : np.sum})
total_price
```

```python
y = res_df['price']
X = res_df.loc[:,['horsepower','engine-size','peak-rpm','bore','length','highway-mpg','curb-weight']]
X.shape
from sklearn.feature_selection import SelectKBest
kbest = SelectKBest(k = 2)
X_new = kbest.fit_transform(X, y)
kbest.get_params()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new, y, random_state = 42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import Lasso
clf = Lasso(alpha = 0.1, max_iter = 10000)
```

```python
clf.fit(X_train, y_train)
```

```python
print(clf.coef_)
print(clf.intercept_)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.linear_model import Lasso
val = [0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4]
for i in val:
    clf = Lasso(alpha = i, max_iter = 10000)
    clf.fit(X_train, y_train)
    print(clf.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators = 100, random_state = 42)
reg.fit(X_train, y_train)
pred = reg.predict(X_test)

from sklearn.metrics import mean_absolute_error
acc = mean_absolute_error(pred, y_test)
print(acc)
```
